```python
class A():
    def __init__(self):
        self.v = 'v'
    
    def f(self):
        print('v' in locals())
        print('self.v' in locals())
        print('v' in globals())
        print('self.v' in globals())
        print(hasattr(self, 'v'))
        
    def f1(a, b=2, self=None):
        print('f1', a, b, self)
        if self: print(self.v)
        print(hasattr(A, 'v'))
        
    def f2(self=None, p=None):
        print('f2', self, isinstance(self, A), p)
        
    def __f3(self): 
        print('f3')
        
    def f4(self):
        print(hasattr(self, '__f3'))
        print(self.__f3)
        print(hasattr(self, '_A__f3'))
        self.__f3()
        
    def f5():
        print('f5')
        
    def f6(self):
        print(b)
        
a = A()
# a.f()
# A.f1(1)
# a.f1(1)
# a.f4()
```

```python
A.f2()
a.f2()
A.f2(1)
a.f2(2)
```

### Iterable, Iterator and Generator

```python
l = [1,2,3]
itt = (_ for _ in l)
print(next(itt))
l.remove(1)
print(next(itt))
```

```python
l1 = [1,2,3]
l1[:0] = [-1]
```

```python
l = [1,2,3]
itt = (_ for _ in l)
print(next(itt))
l[:0] = [None]
l.remove(1)
print(next(itt))
```
