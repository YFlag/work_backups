```python
import numpy as np
import feature_visualization as fv


"""
transformation 1: space bending from 1-d line to 2-d circle spaces.
- t
- f
"""
sampling_magnitude = 50
feature_input_s = fv.sampling(sampling_magnitude, sampling_style='line')
transform_matrix = [
    [lambda x1, x2: np.cos(np.pi*x2 + np.pi), 
     lambda x1, x2: -np.sin(np.pi*x2 + np.pi)]
]

fv.feature_plot(
    feature_input_s.T, 
    fv.transform(feature_input_s, transform_matrix).T, 
    connection_density=1, 
)
```

```python
import numpy as np
import feature_visualization as fv


"""
transformation 2: linear transformation within 2-d spaces.
maybe you can view it as a kind of 'dimension-reducing attack...'
"""
sampling_magnitude = 10
feature_input_s = fv.sampling(sampling_magnitude, sampling_style='square')
labels = np.random.random((sampling_magnitude ** 2,))
transform_matrix = [
    [1, 1],
    [0, 0]
]

""" 'label_method' option: None, 'tripcolor,', 'tricontour', 'tricontourf' """
fv.feature_plot(
    feature_input_s.T, 
    fv.transform(feature_input_s, transform_matrix).T,
    labels=labels,
#     label_method='tripcolor',
    connection_density=1,
    connection_color='b'
)
```

```python
import numpy as np
import feature_visualization as fv


"""
transformation 3: linear transformation within 2-d spaces.
"""
sampling_magnitude = 10
feature_input_s = fv.sampling(sampling_magnitude, sampling_style='square')
labels = np.random.random((sampling_magnitude ** 2,))
transform_matrix = [
    [1, -1],
    [1, 1]
]

""" `label_method` option: 'pcolormesh', 'contour', 'contourf' """
label_method = 'contour'
fv.feature_plot(
    feature_input_s.T.reshape(2, sampling_magnitude, sampling_magnitude), 
    fv.transform(feature_input_s, transform_matrix).T.reshape(
        2, sampling_magnitude, sampling_magnitude
    ), 
    labels=labels.reshape(sampling_magnitude, sampling_magnitude),
    label_method=label_method,   
#     connection_density=0.2
#     line_style='o'
)
```

```python
import numpy as np
import feature_visualization as fv


"""
transformation 3: non-linear transformation within 2-d spaces.
"""
sampling_magnitude = 20
feature_input_s = fv.sampling(sampling_magnitude, sampling_style='square')
labels = np.random.random((sampling_magnitude ** 2))
transform_matrix = [
    [lambda x_1, x_2: x_1 ** 3,
     lambda x_1, x_2: x_1 ** 3 + x_2]
]

""" `label_method` option: 'pcolormesh', 'contour', 'contourf' """
label_method = 'pcolormesh'
fv.feature_plot(
    feature_input_s.T.reshape(2, sampling_magnitude, sampling_magnitude), 
    fv.transform(feature_input_s, transform_matrix).T.reshape(
        2, sampling_magnitude, sampling_magnitude
    ), 
    labels=labels.reshape(sampling_magnitude, sampling_magnitude),
    label_method=label_method, 
    line_style='yo'
)
```

```python
import numpy as np
import feature_visualization as fv


"""
transformation 3: non-linear transformation within 2-d spaces.
"""
sampling_magnitude = 10
feature_input_s = fv.sampling(sampling_magnitude, sampling_style='square')
transform_matrix = [
    [lambda x_1, x_2: np.exp(x_1) / (np.exp(x_1) + np.exp(x_2)),
     lambda x_1, x_2: np.exp(x_2) / (np.exp(x_1) + np.exp(x_2))]
]

fv.feature_plot(
    feature_input_s.T, 
    fv.transform(feature_input_s, transform_matrix).T, 
    connection_density=1,
    connection_color='y'
)
```
